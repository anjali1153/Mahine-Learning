# Defining X and Y 

```python
df['YearsExperience']
```

```python
# Gives output in dataframe/table
df[['YearsExperience']]
```

```python
# In simple linear regression ,only 1 independent column is there but sklearn library always consider that there are multiple x so give it df instead of series
x = df[['YearsExperience']] 
y = df['Salary']
```

```python
# x column must be in 2D format so use double [[]] to convert it into 2D
```

# Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(x,y)           # Training model (calculating m & c values with fit function.)
```

```python
lr.coef_              # Value of slope i.e. m
```

```python
lr.intercept_          # Value of intercept i.e. c
```

```python
m = 9449.96232146
c = 24848
i = 11
m*i+c
```

# Predicting for Actual values

```python
df.head()
```

```python
x 
```

```python
y
```

```python
# Get predicted values on real years exp column for evaluation purpose
ypred = lr.predict(x)
ypred
```

# Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y,ypred)    # Calculating mse value by using mean_squared_error() function
mse
```

```python
rmse = np.sqrt(mse)               # rmse = sqrt(mse)
rmse
```

```python
r2_score = r2_score(y,ypred)         # Calculation r2_score i.e. accuracy of the model
r2_score
```
